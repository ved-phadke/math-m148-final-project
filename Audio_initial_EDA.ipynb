```python
# Convert single-element lists to scalars
df['tempo'] = df['tempo'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 0)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Function to apply PCA on a specific feature column
def apply_pca(features_list, n_components, prefix):
    X = np.array(features_list.tolist())
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)
    pca_columns = [f"{prefix}_pca_{i+1}" for i in range(n_components)]
    return pd.DataFrame(X_pca, columns=pca_columns)

# Apply PCA on mfcc_features, chroma_features, and spectrogram_features
df_mfcc_pca = apply_pca(df["mfcc_features"], n_components=10, prefix="mfcc")
df_chroma_pca = apply_pca(df["chroma_features"], n_components=5, prefix="chroma")
df_spectrogram_pca = apply_pca(df["spectrogram_features"], n_components=20, prefix="spectrogram")

# Concatenate the reduced components back into the original DataFrame
df = pd.concat([df.reset_index(drop=True), df_mfcc_pca, df_chroma_pca, df_spectrogram_pca], axis=1)

# Verify the result
print(df.head())
```

```python
df = df.drop(columns = ['spectrogram_features', 'mfcc_features', 'chroma_features'])
```

```python
# Combine spectrogram PCA columns into a single list column
df["spectrogram_pca"] = df[[f"spectrogram_pca_{i+1}" for i in range(20)]].values.tolist()

# Combine mfcc PCA columns into a single list column
df["mfcc_pca"] = df[[f"mfcc_pca_{i+1}" for i in range(10)]].values.tolist()

# Combine chroma PCA columns into a single list column
df["chroma_pca"] = df[[f"chroma_pca_{i+1}" for i in range(5)]].values.tolist()

# Drop the individual columns if you only want the list columns
df = df.drop(columns=[f"spectrogram_pca_{i+1}" for i in range(20)] +
                      [f"mfcc_pca_{i+1}" for i in range(10)] +
                      [f"chroma_pca_{i+1}" for i in range(5)])
```
