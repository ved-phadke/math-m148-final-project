```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

stimuli_cluster = df.groupby('Stimulus').agg({
    'mean_mfcc': 'mean',
    'spectrogram_mean': 'mean',
    'mean_chroma': 'mean',
    'Chills?': lambda x: (x == 'Yes').mean()  # Proportion of "Yes" for Chills
}).reset_index()

scaler = StandardScaler()
scaled_features = scaler.fit_transform(stimuli_cluster[['mean_mfcc', 'spectrogram_mean', 'mean_chroma']])
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
stimuli_cluster['Cluster'] = kmeans.fit_predict(scaled_features)

plt.figure(figsize=(12, 6))
sns.scatterplot(data=stimuli_cluster, x='mean_mfcc', y='spectrogram_mean', hue='Cluster', palette='coolwarm')
plt.title('Clusters of Stimuli Based on Audio Features')
plt.show()

for cluster in sorted(stimuli_cluster['Cluster'].unique()):
    print(f"Cluster {cluster}:")
    print(stimuli_cluster[stimuli_cluster['Cluster'] == cluster]['Stimulus'].tolist())
    print()
```

cluster 0: mosts likely to have chills
cluster 2: less likely
cluster 1: middle

```python
def pad_or_truncate(arr, target_length=1000):
    if len(arr) >= target_length:
        return arr[:target_length]
    else:
        return np.pad(arr, (0, target_length - len(arr)), mode='constant')

# Apply to each feature column
df_all_features['mfcc_features'] = df_all_features['mfcc_features'].apply(pad_or_truncate)
df_all_features['chroma_features'] = df_all_features['chroma_features'].apply(pad_or_truncate)
df_all_features['spectrogram_features'] = df_all_features['spectrogram_features'].apply(pad_or_truncate)
```
