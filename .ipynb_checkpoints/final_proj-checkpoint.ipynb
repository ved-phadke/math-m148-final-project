```python
transcripts['Stimulus'] = transcripts['Stimulus name'].apply(shorten_title)
transcripts = transcripts.drop('Stimulus name', axis=1)
transcripts.head()
```

```python
counts = chills.groupby('Stimulus')['Chills?'].value_counts().unstack(fill_value=0).reset_index()
counts = counts.merge(transcripts, on='Stimulus', how = 'left')
counts.head()
```

# Converting to a binary variable

```python
counts['chills'] = (counts['Yes'] / (counts['Yes'] + counts['No'])).round()
counts = counts.drop(['Yes','No'], axis=1)
counts.head()
```

# Cleaning the Text

```python
input = list(counts['transcript'])
stop_words = set(stopwords.words('english'))

def clean_text(text):
    if not isinstance(text, str):
       text = str(text)
       
    text = re.sub("\d+", "", text)  # Remove numbers
    text = re.sub(r"[^\w\d'\s]+", " ", text)
    text = text.lower()
    words = word_tokenize(text)  # Tokenization
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]  # Lemmatization & stopword removal
    return " ".join(words)

input = [clean_text(i) for i in tqdm(input)]
counts["clean_text"] = input
```

```python
counts.head()
```

# Embeddings

```python
from gensim.models import Word2Vec
import numpy as np
from sklearn.decomposition import PCA

text_data = [sentence.split() for sentence in counts["clean_text"]]

model = Word2Vec(sentences=text_data, vector_size=100, window=5, min_count=1, workers=4)

def get_sentence_embedding(sentence):
    words = sentence.split()
    word_vectors = [model.wv[word] for word in words if word in model.wv]
    return np.mean(word_vectors, axis=0) if word_vectors else np.zeros(model.vector_size)

counts["embeddings"] = counts["clean_text"].apply(get_sentence_embedding)
```

# TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=300)  # Adjust max_features as needed

tfidf_matrix = tfidf_vectorizer.fit_transform(counts["clean_text"])

counts["tfidf_vector"] = [row.toarray().flatten() for row in tfidf_matrix]
```

# BERT

```python
df = counts[['chills', 'clean_text']]

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(f"Using device: {device}")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertModel, get_scheduler

class BERTProbabilityModel(nn.Module):
    def __init__(self, model_name="bert-base-uncased"):
        super(BERTProbabilityModel, self).__init__()
        self.bert = BertModel.from_pretrained(model_name)
        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(self.bert.config.hidden_size, 1)  # Single neuron for probability
        self.sigmoid = nn.Sigmoid()  # Convert logits to probability

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output  # [CLS] token representation
        x = self.dropout(pooled_output)
        prob = self.sigmoid(self.fc(x))  # Outputs probability
        return prob


tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
class ProbabilityDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        encodings = tokenizer(texts, padding=True, truncation=True, max_length=max_length, return_tensors="pt")

        self.input_ids = encodings["input_ids"]
        self.attention_mask = encodings["attention_mask"]
        self.labels = torch.tensor(labels, dtype=torch.float32)  # Ensure label is a tensor

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return {
            "input_ids": self.input_ids[idx],
            "attention_mask": self.attention_mask[idx],
            "label": self.labels[idx]
        }


texts = df["clean_text"].tolist()
labels = df["chills"].tolist() 
```

```python
from sklearn.model_selection import train_test_split

train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    texts, labels, test_size=0.2, random_state=1, stratify=labels)

val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts, temp_labels, test_size = 0.5, random_state=1, stratify=temp_labels)



# Create datasets
train_dataset = ProbabilityDataset(train_texts, train_labels, tokenizer)
val_dataset = ProbabilityDataset(val_texts, val_labels, tokenizer)

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model = BERTProbabilityModel().to(device)
```

```python
# Define loss and optimizer
criterion = nn.BCELoss()  # Binary cross-entropy loss for classification
optimizer = optim.AdamW(model.parameters(), lr=2e-5)

# Learning rate scheduler (optional)
num_training_steps = len(train_loader) * 5  # 5 epochs
lr_scheduler = get_scheduler("linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps)
```

```python
# Training loop
num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    correct_predictions = 0

    for batch in train_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["label"].to(device).unsqueeze(1)  # Reshape for BCELoss
        
        optimizer.zero_grad()
        
        outputs = model(input_ids, attention_mask)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        lr_scheduler.step()

        total_loss += loss.item()
        
        # Calculate accuracy
        preds = (outputs >= 0.5).float()
        correct_predictions += (preds == labels).sum().item()

    avg_loss = total_loss / len(train_loader)
    accuracy = correct_predictions / len(train_dataset)
    
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}, Accuracy: {accuracy:.4f}")

# Save trained model
torch.save(model.state_dict(), "bert_chills_classifier.pth")
print("Model saved!")

# Evaluate on validation set
model.eval()
val_loss = 0
correct_predictions = 0

with torch.no_grad():
    for batch in val_loader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["label"].to(device).unsqueeze(1)
        
        outputs = model(input_ids, attention_mask)
        loss = criterion(outputs, labels)
        
        val_loss += loss.item()
        
        preds = (outputs >= 0.5).float()
        correct_predictions += (preds == labels).sum().item()

avg_val_loss = val_loss / len(val_loader)
val_accuracy = correct_predictions / len(val_dataset)

print(f"Validation Loss: {avg_val_loss:.4f}, Accuracy: {val_accuracy:.4f}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Convert embeddings and TF-IDF from lists to numpy arrays
X_embeddings = np.vstack(counts["embeddings"].values)  # Shape: (n_samples, embedding_dim)
X_tfidf = np.vstack(counts["tfidf_vector"].values)     # Shape: (n_samples, tfidf_dim)
X_sentiment = counts[["sentiment"]].values             # Shape: (n_samples, 1)

# Combine all features into one matrix
X = np.hstack([X_embeddings, X_tfidf, X_sentiment])  # Final shape: (n_samples, total_features)

# Define target variables
y_yes = counts["Yes"].values  
y_no = counts["No"].values  

X_train, X_test, y_train_yes, y_test_yes = train_test_split(X, y_yes, test_size=0.2, random_state=1)
X_train, X_test, y_train_no, y_test_no = train_test_split(X, y_no, test_size=0.2, random_state=1)

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

model_yes = RandomForestRegressor(n_estimators=100, random_state=1)
model_yes.fit(X_train, y_train_yes)

model_no = RandomForestRegressor(n_estimators=100, random_state=1)
model_no.fit(X_train, y_train_no)

y_pred_yes = model_yes.predict(X_test)
y_pred_no = model_no.predict(X_test)

mse_yes = mean_squared_error(y_test_yes, y_pred_yes)
mse_no = mean_squared_error(y_test_no, y_pred_no)

print(f"MSE for 'Yes' count prediction: {mse_yes}")
print(f"MSE for 'No' count prediction: {mse_no}")
```
